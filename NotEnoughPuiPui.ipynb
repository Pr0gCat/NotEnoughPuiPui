# GAN出更多天竺鼠車車圖片

```python

import torch
import torch.nn as nn
import torch.optim as optimizers
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from torchvision.utils import make_grid
import random
import numpy as np
from torchsummary import summary

import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML, display
import ipywidgets as widgets
```

```python
MODEL_2_USE = 'wgan'
if MODEL_2_USE == 'dcgan':
    from dcgan import NEPPGenerator
    g = NEPPGenerator().eval()
    g.load_state_dict(torch.load('dcgan_models/g_net.pth'))
elif MODEL_2_USE == 'wgan':
    from wgan import NEPPGenerator
    g = NEPPGenerator().eval()
    g.load_state_dict(torch.load('wgan_models/g_net.pth'))
    
```

```python
def gen_puipui(generator, n=8):
    noise = torch.randn(n, 100, 1, 1).cuda()
    raw_puipui = generator(noise).detach().cpu()
    return np.transpose(make_grid(raw_puipui, padding=2, normalize=True),(1,2,0)).numpy()
```

```python
manualSeed = 666
random.seed(manualSeed)
torch.manual_seed(manualSeed)
plt.imshow(gen_puipui(g, 32))
```
